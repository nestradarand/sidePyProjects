```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('hci_text.csv')
```

```python
df['responses'] = df['responses'].str.lower()
df['responses'] = df['responses'].str.strip('"')
```

```python
one_string = " ".join(review for review in df['responses'])
```

```python
stopwords = STOPWORDS
stopwords.update(['click','use','incident'])
```

```python
words = WordCloud(max_font_size = 50, max_words = 8,stopwords=stopwords,
background_color='white').generate(one_string)
```

```python
plt.imshow(words, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
words.to_file("hci_cloud.png")
```
